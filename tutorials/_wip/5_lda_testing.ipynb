```python
import shnitsel as st
import shnitsel.xarray
```

```python
dt = st.read("../test_data/shnitsel/traj_I02.nc")
dt
```

```python
from shnitsel.clean import sanity_check

cleaned_dt = sanity_check(dt)
cleaned_dt
```

```python
from shnitsel.filtering import StructureSelection

struct_sel = StructureSelection.init_from_dataset(cleaned_dt['I02/0/data'], default_selection=['bonds','angles','dihedrals','pyramids']).non_redundant()
struct_sel
```

```python
from shnitsel.geo.geocalc import get_bats

bats = get_bats(cleaned_dt, struct_sel, signed=True, angles='deg')
bats
```

```python
from shnitsel.analyze.lda import lda

lda_res = lda(bats, categories=cleaned_dt.astate, dim='descriptor')
```

```python
for p, x in lda_res.explain_loadings().collect_data(True):
    print(f"LDA for {p}:")
    print(x)
```

```python
from shnitsel.vis.plot.biplot import biplot_kde

biplot_kde(cleaned_dt, pca_data=lda_res, structure_selection=struct_sel, scatter_color_property='state')
```
